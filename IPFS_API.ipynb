```python
import tweepy
#import sqlite3
import time
import ipfsapi

#db = sqlite3.connect('data/MyDB.db')
# Get a cursor object
#cursor = db.cursor()
#cursor.execute('''CREATE TABLE IF NOT EXISTS MyTable(id INTEGER PRIMARY KEY, name TEXT, geo TEXT, image TEXT, source TEXT, timestamp TEXT, text TEXT, rt INTEGER)''')
#db.commit()

consumer_key = str(input("consumer_key:"))
consumer_secret =str(input("consumer_secret:"))

key = str(input("key:"))
secret =str(input("secret:"))

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(key, secret)

api = tweepy.API(auth)

search=str(input("hashtag:"))

c = tweepy.Cursor(api.search,
                       q=search,
                       include_entities=True).items(10)
tweets=[]

while True:
    try:
        tweet = c.next()
        tweets.append(tweet)
        print tweet
        # Insert into db
    except tweepy.TweepError:
        time.sleep(60 * 15)
        continue
    except StopIteration:
        break


```

```python
#write tweets to a file

thefile = open('test2.txt', 'w')
for item in tweets:
  thefile.write("%s\n" % item)
```
